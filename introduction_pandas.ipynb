### Importando o pandas

```python
import pandas as pd

# O pd é um apelido para facilitar a escrita.
```

### Criando um dataframe a partir de um dicionário

```python
# O pandas trabalha com dataframes (tabela no python)
# dataframe = pd.DataFrame() cria um dataframe vazio
venda = {'data': ['15/02/2021', '16/02/2021'],
         'valor': [500, 300],
         'produto': ['feijao', 'arroz'],
         'qtde': [50, 70],
        }
# Você pode criar um dataframe a partir de um dicionário em python
vendas_df = pd.DataFrame(venda)
```

### Visualização dos Dados
    - print
    - display

```python
display(vendas_df)
```

### Importando arquivos e bases de dados

```python
vendas_df = pd.read_excel("Vendas.xlsx")
display(vendas_df)
```

### Resumos de Visualização de Dados simples e úteis
    - head
    - shape
    - describe

```python
# outros metodos para visualizar os dados

display(vendas_df.head()) #Mostra as 5 primeiras linhas ou a quantidade que você desejar, basta colocar a quantidade de linhas no argumento da função head().
print(vendas_df.shape) # Mostra quantas linhas e quantas colunas temos na tabel de forma instantanea.
display(vendas_df.describe())
#O describe me dá dados estatisticos da minha tabela.. tudo calculado já! Ele me da média, mediana etc..
```

### Pegar 1 coluna (e os pd.Series)

```python
# sempre que eu quiser pegar uma unica coluna da minha tabela
# para pegar mais de um valor basta eu usar mais um colchetes e colocar mais uma coluna.
produtos = vendas_df[['Produto', 'Quantidade']]
display(produtos)
```

### .loc, um método muito importante
    - Pegar 1 linha
    - Pegar linhas de acordo com alguma condição
    - Pegar linhas e colunas específicas
    - Pegar 1 valor específico

```python
# pegar uma linha 
display(vendas_df.loc[1:5])

#pegar linhas que correspondem a uma condição, ou seja, eu vou ter todas as linhas em que, e todas as colunas tal que uma coluna específica tem um valor específico

display(vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping'])

#pegar  várias linhas e colunas usando o loc

vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ["ID Loja", "Produto", "Quantidade"]]
display(vendas_norteshopping_df)

#Pegar um cara específico da minha tabela 
print(vendas_df.loc[1, 'Produto']) # me da um unico elemento

```

### Adicionar 1 coluna

```python
# A partir de uma coluna que já existe 
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
# display(vendas_df)
# cria uma coluna a partir de uma que já existe


# Criar uma coluna com valor padrão
vendas_df.loc[:, 'Imposto'] = 0 # Essa maneira demora menos a processar
display(vendas_df) 
```

### Adicionar 1 linha
    - Linhas de um complemento da base de dados

```python
vendas_dez_df = pd.read_excel("Vendas - Dez.xlsx")
# display(vendas_dez_df)

vendas_df = pd.concat([vendas_df, vendas_dez_df], ignore_index=True)
display(vendas_df)
```

### Valores Vazios
    - Deletar linhas/colunas vazias
    - Deletar linhas que possuem valores vazios
    - Preencher valores vazios (média e último valor)

```python
# Deletar linhas e colunas completamente vazias 

vendas_df = vendas_df.dropna(how = 'all', axis=1) #exclui todos os valores vazios das colunas, axis=0 exclui das linhas.

# deletar linhas que possuem pelo menos 1 valor vazio
vendas_df = vendas_df.dropna()

# preencher valores vazios
#preencher com a média das colunas
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean())
display(vendas_df)

# Preencher com o ultimo valor
vendas_df = vendas_df.ffill()
display(vendas_df)

```

### Calcular Indicadores
    - Groupby
    - Value Counts

```python
# value counts 
transacoes_produto = vendas_df['Produto'].value_counts()
display(transacoes_produto)

# goup by 
faturamento_produto = vendas_df.groupby('Produto').sum(numeric_only=True)
display(faturamento_produto)

#Podemos filtrar as colunas
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum(numeric_only=True)
display(faturamento_produto)

```

### Mesclar 2 dataframes (Procurar informações de um dataframe em outro)

```python
gerentes_df = pd.read_excel('Gerentes.xlsx')
# display(gerentes_df)

vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
